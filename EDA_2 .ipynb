## Scenarios for Scaling Techniques

**Standard Scaling (StandardScaler):**
- Preferred when data has **outliers**.  
- Useful for models that assume a **normal (Gaussian) distribution**, such as Linear Regression, Logistic Regression, and SVM.  
- Scales features to mean = 0 and standard deviation = 1.  

**Min-Max Scaling (MinMaxScaler):**
- Preferred when you want features in a **fixed range** (commonly 0 to 1).  
- Works well for algorithms that use **distance or gradients**, like k-Nearest Neighbors and Neural Networks.  
- Not good if data has strong outliers, as they distort the scale.  



2. Encoding Techniques:

```python
categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical columns:")
display(categorical_cols)

# Separate categorical columns based on the number of unique values
low_cardinality_cols = [col for col in categorical_cols if df[col].nunique() < 5]
high_cardinality_cols = [col for col in categorical_cols if df[col].nunique() >= 5]

print("\nCategorical columns with less than 5 unique values:")
display(low_cardinality_cols)

print("\nCategorical columns with 5 or more unique values:")
display(high_cardinality_cols)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Apply One-Hot Encoding to low cardinality columns
one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
df_encoded_low = one_hot_encoder.fit_transform(df[low_cardinality_cols])

# Create a DataFrame with the one-hot encoded columns
df_encoded_low = pd.DataFrame(df_encoded_low, columns=one_hot_encoder.get_feature_names_out(low_cardinality_cols))

print("DataFrame after One-Hot Encoding (low cardinality columns):")
display(df_encoded_low.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Apply Label Encoding to high cardinality columns
label_encoders = {}
df_encoded_high = df[high_cardinality_cols].copy()

for col in high_cardinality_cols:
    label_encoders[col] = LabelEncoder()
    df_encoded_high[col] = label_encoders[col].fit_transform(df_encoded_high[col])

print("DataFrame after Label Encoding (high cardinality columns):")
display(df_encoded_high.head())
```

## One-Hot Encoding vs Label Encoding

**One-Hot Encoding**
- **Pros:**  
  - No order/priority assigned to categories.  
  - Works well for algorithms that cant handle categorical values directly.  
- **Cons:**  
  - Increases feature space (many columns if categories are large).  
  - Can lead to sparse data.  

**Label Encoding**
- **Pros:**  
  - Simple and memory-efficient (only one column).  
  - Useful when categories have a natural order (e.g., small, medium, large).  
- **Cons:**  
  - Assigns numbers that may imply an order when none exists.  
  - Can mislead algorithms into thinking one category is greater than another.


3. Feature Engineering:

```python
# 1. Create 'capital_total' feature
df['capital_total'] = df['capital_gain'] + df['capital_loss']

# 2. Create 'hours_per_week_binned' feature by binning 'hours_per_week'
bins = [0, 20, 40, 60, 100]  # Define bin edges
labels = ['Part-time', 'Full-time', 'Overtime', 'Very High'] # Define labels for bins
df['hours_per_week_binned'] = pd.cut(df['hours_per_week'], bins=bins, labels=labels, right=False)

print("DataFrame with new features:")
display(df[['capital_gain', 'capital_loss', 'capital_total', 'hours_per_week', 'hours_per_week_binned']].head())
```

## Rationale for New Features

**capital_total:** Created by summing `capital_gain` and `capital_loss`. This feature provides a single metric representing the net change in an individual's capital. Capital gains and losses can significantly impact an individual's financial status, and combining them into a single feature might capture this aspect more effectively than treating them separately, especially if the model benefits from a consolidated financial indicator.

**hours_per_week_binned:** Created by categorizing the `hours_per_week` into meaningful bins: 'Part-time' (0-19 hours), 'Full-time' (20-39 hours), 'Overtime' (40-59 hours), and 'Very High' (60+ hours). Binning can help to:
1. Handle potential non-linear relationships between hours worked and the target variable (income).
2. Reduce the impact of outliers in `hours_per_week`.
3. Make the feature more interpretable by grouping individuals into standard work categories.


## Rationale for Log Transformation

Log transformation (specifically using `np.log1p` which is $\log(1+x)$) was applied to the highly skewed numerical features: `fnlwgt`, `capital_gain`, `capital_loss`, and `capital_total`.

The rationale for this transformation is:
1.  **Reduce Skewness:** Highly skewed data can violate assumptions of some machine learning models (like linear models) and negatively impact their performance. Log transformation is a common technique to reduce positive skewness and make the distribution more symmetric, closer to a normal distribution.
2.  **Stabilize Variance:** For features with values spanning several orders of magnitude (common in skewed data), log transformation can help to stabilize the variance, making patterns in the data more consistent across different ranges of the feature.
3.  **Handle Zero Values:** Using `np.log1p` is crucial for features like `capital_gain`, `capital_loss`, and `capital_total` which contain zero values. Applying a standard `np.log` to zero would result in negative infinity, which is not suitable for analysis. `log1p(x)` is equivalent to `log(1+x)`, so `log1p(0)` is `log(1)` which equals 0, allowing the transformation to be applied across all values, including zero.

As seen in the output, the skewness of the transformed features (`fnlwgt_log`, `capital_gain_log`, `capital_loss_log`, `capital_total_log`) is significantly reduced compared to their original counterparts, although some still exhibit some level of skewness. This transformation makes these features more suitable for use in various machine learning algorithms.

## Relationship Between Features and Correlation Matrix

When analyzing a dataset, we often want to understand how features are related:

- **Feature Relationships:**
  - By exploring scatter plots, pair plots, or association rules, we can see patterns like linear, non-linear, or categorical relationships.
  - For example, sales might increase with advertising spend, or height might relate to weight.

- **Correlation Matrix:**
  - A correlation matrix shows the **strength and direction of linear relationships** between numerical features.
  - Values close to +1 indicate strong positive correlation, values close to -1 indicate strong negative correlation, and values near 0 mean weak/no linear relationship.

- **Comparison:**
  - Feature exploration (plots, patterns) gives a **visual and broader view**, including non-linear and categorical relationships.
  - The correlation matrix gives a **quantitative summary** but only measures linear associations.
  - Together, they provide a more complete understanding of how features interact.


```python
import math
import numpy as np
import pandas as pd
from typing import Optional, Dict, Any
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.preprocessing import OrdinalEncoder
from sklearn.metrics import log_loss, mean_absolute_error
import seaborn as sns
import matplotlib.pyplot as plt


def _is_categorical(series: pd.Series, max_unique_ratio: float = 0.05, max_unique_abs: int = 20) -> bool:
    if pd.api.types.is_object_dtype(series) or pd.api.types.is_categorical_dtype(series):
        return True
    if pd.api.types.is_bool_dtype(series):
        return True
    if pd.api.types.is_numeric_dtype(series):
        nunique = series.dropna().nunique()
        if nunique <= max_unique_abs or (nunique / max(1, series.dropna().shape[0])) <= max_unique_ratio:
            return True
    return False


def _prepare_feature(x: pd.Series) -> np.ndarray:
    x_ = x.to_numpy().reshape(-1, 1)
    if _is_categorical(x):
        enc = OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1)
        return enc.fit_transform(x_)
    return x_


def _prepare_target_classifier(y: pd.Series) -> (np.ndarray, Dict[Any, int]):
    y_ = y.astype("category")
    mapping = {cat: i for i, cat in enumerate(y_.cat.categories)}
    y_idx = y_.cat.codes.replace(-1, np.nan)
    return y_idx.to_numpy(), mapping


def _baseline_logloss(y: np.ndarray) -> float:
    # Baseline = predict marginal class probabilities for every row
    # y are integer class indices with possible NaNs; drop NaNs for baseline fit
    y_non_nan = y[~np.isnan(y)].astype(int)
    class_counts = np.bincount(y_non_nan)
    probs = class_counts / class_counts.sum()
    eps = 1e-15
    probs = np.clip(probs, eps, 1 - eps)
    return -np.sum(probs * np.log(probs))


def _baseline_mae(y: np.ndarray) -> float:
    y_non_nan = y[~np.isnan(y)]
    if y_non_nan.size == 0:
        return np.nan
    median_value = np.median(y_non_nan)
    return np.mean(np.abs(y_non_nan - median_value))


def pps_single_feature(df: pd.DataFrame, feature: str, target: str, n_splits: int = 5, random_state: int = 42) -> Optional[float]:
    if feature == target:
        return 0.0
    x = df[feature]
    y = df[target]

    valid_mask = ~(x.isna() | y.isna())
    x = x[valid_mask]
    y = y[valid_mask]
    if x.shape[0] < 30:
        return 0.0

    X = _prepare_feature(x)

    if _is_categorical(y):
        y_idx, _ = _prepare_target_classifier(y)
        if np.unique(y_idx[~np.isnan(y_idx)]).size <= 1:
            return 0.0
        kf = KFold(n_splits=min(n_splits, np.unique(y_idx[~np.isnan(y_idx)]).size, x.shape[0]), shuffle=True, random_state=random_state)
        model = DecisionTreeClassifier(max_depth=4, random_state=random_state)
        logloss_values = []
        for train_idx, test_idx in kf.split(X):
            X_train, X_test = X[train_idx], X[test_idx]
            y_train, y_test = y_idx[train_idx].astype(float), y_idx[test_idx].astype(float)

            mask_train = ~np.isnan(y_train)
            mask_test = ~np.isnan(y_test)
            if mask_train.sum() < 2 or mask_test.sum() == 0:
                continue

            model.fit(X_train[mask_train], y_train[mask_train].astype(int))
            probs = model.predict_proba(X_test[mask_test])

            n_classes = int(np.nanmax(y_idx)) + 1
            if probs.shape[1] < n_classes:
                pad = np.zeros((probs.shape[0], n_classes - probs.shape[1]))
                probs = np.hstack([probs, pad])

            fold_logloss = log_loss(y_test[mask_test].astype(int), probs, labels=list(range(n_classes)))
            logloss_values.append(fold_logloss)

        if len(logloss_values) == 0:
            return 0.0

        model_logloss = float(np.mean(logloss_values))
        base_logloss = float(_baseline_logloss(y_idx))
        if not math.isfinite(model_logloss) or not math.isfinite(base_logloss) or base_logloss <= 0:
            return 0.0
        score = 1.0 - (model_logloss / base_logloss)
        return float(np.clip(score, 0.0, 1.0))
    else:
        y_num = pd.to_numeric(y, errors="coerce").to_numpy()
        if np.nanstd(y_num) == 0:
            return 0.0

        kf = KFold(n_splits=min(n_splits, x.shape[0]), shuffle=True, random_state=random_state)
        model = DecisionTreeRegressor(max_depth=4, random_state=random_state)
        mae_values = []
        for train_idx, test_idx in kf.split(X):
            X_train, X_test = X[train_idx], X[test_idx]
            y_train, y_test = y_num[train_idx], y_num[test_idx]

            mask_train = ~np.isnan(y_train)
            mask_test = ~np.isnan(y_test)
            if mask_train.sum() < 2 or mask_test.sum() == 0:
                continue

            model.fit(X_train[mask_train], y_train[mask_train])
            y_pred = model.predict(X_test[mask_test])
            mae_values.append(mean_absolute_error(y_test[mask_test], y_pred))

        if len(mae_values) == 0:
            return 0.0

        model_mae = float(np.mean(mae_values))
        base_mae = float(_baseline_mae(y_num))
        if not math.isfinite(model_mae) or not math.isfinite(base_mae) or base_mae <= 0:
            return 0.0
        score = 1.0 - (model_mae / base_mae)
        return float(np.clip(score, 0.0, 1.0))


def pps_matrix(df: pd.DataFrame, columns: Optional[list] = None, cap_columns: Optional[int] = 30, random_state: int = 42) -> pd.DataFrame:
    cols = list(columns) if columns is not None else list(df.columns)
    if cap_columns is not None and len(cols) > cap_columns:
        rng = np.random.default_rng(random_state)
        cols = list(rng.choice(cols, size=cap_columns, replace=False))
    mat = pd.DataFrame(0.0, index=cols, columns=cols, dtype=float)
    for target in cols:
        for feature in cols:
            try:
                score = pps_single_feature(df, feature, target, random_state=random_state)
                mat.loc[target, feature] = score if score is not None else 0.0
            except Exception:
                mat.loc[target, feature] = 0.0
    return mat


def plot_heatmap(matrix: pd.DataFrame, title: str, cmap: str = "viridis"):
    plt.figure(
        figsize=(min(1.5 * len(matrix.columns), 18), min(1.2 * len(matrix.index), 14))
    )
    sns.heatmap(matrix, annot=False, cmap=cmap, square=True, cbar=True)
    plt.title(title)
    plt.tight_layout()
    plt.show()
```
